```python
#from IPython.display import display, HTML
region = input("Enter the region: ")
x = df_merged[df_merged['Region'] == region]
if not x.empty:
    ph_range = x['pH Range'].values[0]
    print("The pH Range of region", region, "is:", ph_range)
else:
    print("No pH range information found for the specified region:", region)

x = df_merged[df_merged['Region'] == region]
if not x.empty:
    Avg_Nitrogen = x['Avg. Nitrogen Percent'].values[0]
    print("The Average Nitrogen Percent of", region, "is:", Avg_Nitrogen)
else:
    print("No Nitrogen percent information found for the specified region:", region)
    
x = df_merged[df_merged['Region'] == region]
if not x.empty:
    Avg_Carbon = x['Avg. Carbon Percent'].values[0]
    print("The Average Carbon Percent of", region, "is:", Avg_Carbon)
else:
    print("No Carbon percent information found for the specified region:", region)
    
x = df_merged[df_merged['Region'] == region]
if not x.empty:
    Avg_Phosph = x['Avg. Phosphorus(mg/kg)'].values[0]
    print("The Average Phosphorus content of", region, "is:", Avg_Phosph)
else:
    print("No Phosphorus information found for the specified region:", region)
    
x = df_merged[df_merged['Region'] == region]
if not x.empty:
    hum = x['Humidity'].values[0]
    print("The Humidity of", region, "is:", hum)
else:
    print("No Humidity information found for the specified region:", region)

x = df_merged[df_merged['Region'] == region]
if not x.empty:
    temp = x['Avg. Temperature(°C )'].values[0]
    print("The Average Temperature of", region, "is:", temp)
else:
    print("No Temperature information found for the specified region:", region)

x = df_merged[df_merged['Region'] == region]
if not x.empty:
    rain = x['Rainfall_(mm)'].values[0]
    print("The Rainfall of", region, "is:", rain)
else:
    print("No Rainfall information found for the specified region:", region)
    
    
crop=input("Enter the Season to display the favourable crops for that season:  ")
if crop=='Summer Season':

    x = df_merged[df_merged['Region'] == region]

    if not x.empty:
    
        Summer = x['Summer Season Crops'].values[0]
        print("The favourable crops of region", region, " foe this season is:", Summer)

    else:
        print("No crops information found for the specified region:", region)
elif crop=='Winter Season':
    x = df_merged[df_merged['Region'] == region]

    if not x.empty:
    
        winter = x['Winter Season'].values[0]
        print("The favourable crops of region", region, " for this season is:", winter)

    else:
        print("No crops information found for the specified region:", region)
        
elif crop=='Monsoon Season':
    x = df_merged[df_merged['Region'] == region]

    if not x.empty:
    
        monsoon = x['Monsoon Season'].values[0]
        print("The favourable crops of region", region, "is:", monsoon)

    else:
        print("No crops information found for the specified region:", region)
else:
    print("Enter Correct Season")
```
